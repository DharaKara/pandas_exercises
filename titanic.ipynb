# process
- cleaning
- distribution and standardizing

```python
feature_df = titanic_df.loc[:, ['Age', 'Pclass', 'Fare']]
feature_df.head(10)
target_df = titanic_df[['Survived']]
target_df.head(10)

feature_df.isnull().sum()
```

```python
# impute missing value - mean
# pip install scikit-learn
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy="mean")
feautures_imputed_df = imputer.fit_transform(feature_df)
feautures_imputed_df
# 2. cleaning done
```

```python
# 3. standardization - done
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
feautures_scaled = scaler.fit_transform(feautures_imputed_df)
feautures_scaled
```

```python
pd.DataFrame(feautures_imputed_df).describe()
```

```python
(7.25-32.204208)/49.693429 
# (x - mean)/std
```

```python
# 4. feauture engineering
# - bucketing 
# - encoding (string) -> (number)
```

```python
# modelling
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(feautures_scaled, target_df, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train) # training done

# predictions
predict = model.predict(X_test)
predict # model predict
y_test # actual output
```

```python
print(len(X_train), len(X_test))
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, predict)
print(f"Model accuracy: {accuracy:.2%}")
```
